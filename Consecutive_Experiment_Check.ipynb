```python
#Open the file
import pandas as pd 
filepath = "./data/masked_sentences_bert.txt"
df = pd.read_table(filepath, header = None)
df.columns = ["Score","Year","PMCID","Gene","Experiment","Sentence","Masked"]
grouped = df.groupby('PMCID')
sent_ids = []

# Loop through each PMCID group
for _, group in grouped:
    sentences = group["Sentence"].tolist()
    sent_id = 1
    sent_ids.append(sent_id)
    
    # Check for consecutive sentences
    for i in range(1, len(sentences)):
        if sentences[i] != sentences[i-1]:
            sent_id += 1
        sent_ids.append(sent_id)

# Add as a new column
df["SentID"] = sent_ids
```

```python
import random
import pandas as pd
random.seed(100)

pairs = []
while len(pairs) < 10:  #10 pairs for demonstration
    idx = random.randint(0, len(df) - 2)
    row1, row2 = df.iloc[idx], df.iloc[idx+1]
    
    # Check if rows have the same PMCID and different sentences
    if row1["PMCID"] == row2["PMCID"] and row1["Sentence"] != row2["Sentence"]:
        pairs.append((row1["PMCID"].split("-")[0], row1["Sentence"], row2["Sentence"], row1["Gene"], row2["Gene"]))

pairs_df = pd.DataFrame(pairs, columns=["PMCID", "Sentence1", "Sentence2", "Gene1", "Gene2"])
pairs_df.to_csv("./data/pairs.csv", index=False)
```
